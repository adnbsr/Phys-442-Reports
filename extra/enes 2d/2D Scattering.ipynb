```python

def my_error(x,y,sy):
    x=array(x) 
    y=array(y) 
    sy=array(sy)
    x=x*2*pi/360.0
    x=sin(x/2)
    
    S = sum(1 / sy**2)
    Sx = sum(x / sy**2)
    Sy = sum(y / sy**2)
    Sxx = sum(x**2 / sy**2)
    Sxy = sum(x*y / sy**2)
    delta = S*Sxx - Sx**2
    n = (Sxx*Sy - Sx*Sxy) / delta 
    m = (S*Sxy - Sx*Sy) / delta
    sn = sqrt(Sxx / delta) 
    sm = sqrt(S / delta)
    errorbar(x,y,yerr=sy,fmt='o')
    xx = arange(0,1.01,0.01) 
    yy = n + m*xx
    plot(xx, yy,'r-')

    xlabel('$sin({theta/2})$')
    ylabel('dN')

    savefig('error.png')
    #show()
    return m,sm
```
